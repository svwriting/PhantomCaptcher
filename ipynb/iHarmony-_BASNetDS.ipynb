```
from IPython.display import clear_output
from PIL import Image
import numpy as np
import os
import random

iHarmony_dir_path="/content/drive/Shareddrives/幻影捕手/AI班團專_幻影捕手/資料集/Improving_the_Harmony_of_the_Composite_Image/iHarmony4"
BASNetDS_dir_path="/content/drive/Shareddrives/幻影捕手/AI班團專_幻影捕手/資料集/BASNetDS"

def SAVEimg(pic_,filename,rcm=''):
    # pic_.save(f"{BASNetDS_dir_path}/im_aug/{filename}")
    if rcm=='rm':
      pic_.save(f"{BASNetDS_dir_path}/im_aug_rm/{filename}")
    if rcm=='cm':
      pic_.save(f"{BASNetDS_dir_path}/im_aug_cm/{filename}")
def SAVEmask(pic_,filename,rcm=''):
    # pic_.save(f"{BASNetDS_dir_path}/gt_aug/{filename}")
    if rcm=='rm':
      pic_.save(f"{BASNetDS_dir_path}/gt_aug_rm/{filename}")
    if rcm=='cm':
      pic_.save(f"{BASNetDS_dir_path}/gt_aug_cm/{filename}")

dirlist_iHarmony=[ item for item in os.listdir(iHarmony_dir_path) if '.' not in item]
dirlist_HAdobe5k={ f"{iHarmony_dir_path}/{dirlist_iHarmony[0]}":[item for item in os.listdir(f"{iHarmony_dir_path}/{dirlist_iHarmony[0]}") if '.' not in item]}
dirlist_HCOCO={ f"{iHarmony_dir_path}/{dirlist_iHarmony[1]}":[item for item in os.listdir(f"{iHarmony_dir_path}/{dirlist_iHarmony[1]}") if '.' not in item]}
dirlist_HFlickr={ f"{iHarmony_dir_path}/{dirlist_iHarmony[2]}":[item for item in os.listdir(f"{iHarmony_dir_path}/{dirlist_iHarmony[2]}") if '.' not in item]}
dirlist_Hday2night={ f"{iHarmony_dir_path}/{dirlist_iHarmony[3]}":[item for item in os.listdir(f"{iHarmony_dir_path}/{dirlist_iHarmony[3]}") if '.' not in item]}

def subDir(dict_):
    newdict={}
    key_=list(dict_.keys())[0]
    newdict_={}
    for dir_ in dict_[key_]:
        if 'real' in dir_:
            newdict_['real']=f"{key_}/{dir_}"
        elif 'mask' in dir_:
            newdict_['mask']=f"{key_}/{dir_}"
        elif 'composite' in dir_:
            newdict_['composite']=f"{key_}/{dir_}"
    newdict[key_]=newdict_
    return newdict

def run(dirlist_):
    dirlist_=subDir(dirlist_)
    key_=list(dirlist_.keys())[0]
    dirname_=key_[key_.rfind('/')+1:]
    dict_=dirlist_[key_]
    real_dir=dict_['real']
    mask_dir=dict_['mask']
    composite_dir=dict_['composite']


    # len_=len(os.listdir(mask_dir))
    # count_=0
    # for filename_ in os.listdir(mask_dir):
    #     if os.path.exists(f"{BASNetDS_dir_path}/gt_aug_rm/{filename_.split('.')[0]}.png"):
    #         count_+=1
    #         print(f'{dirname_}-rm-({count_}/{len_})',filename_.split('.')[0])
    #         continue
    #     filename_m=filename_
    #     try:
    #       mask_=Image.open(f"{mask_dir}/{filename_}")
    #     except:
    #       continue
    #     filename_=filename_.split('.')[0]
    #     filename_m=filename_m[:filename_m.find('_')]
    #     img_=Image.open(f"{real_dir}/{filename_m}.jpg")
    #     array_=np.array(mask_)
    #     min_x,min_y,max_x,max_y=10000,10000,0,0
    #     for i in range(array_.shape[0]):
    #         for j in range(array_.shape[1]):
    #             if array_[i][j][0] != 0:
    #                 min_x,min_y,max_x,max_y=min(min_x,i),min(min_y,j),max(max_x,i),max(max_y,j)
        
    #     w_=(max_y-min_y+1)//10
    #     h_=(max_x-min_x+1)//10
    #     min_y-=random.randint(0,w_) # 左
    #     min_y=0 if min_y<0 else min_y
    #     min_x-=random.randint(0,h_) # 上
    #     min_x=0 if min_x<0 else min_x
    #     max_y+=random.randint(0,w_) # 右
    #     max_y=array_.shape[1] if max_y>array_.shape[1] else max_y
    #     max_x+=random.randint(0,h_) # 下
    #     max_x=array_.shape[0] if max_x>array_.shape[0] else max_x

    #     mask_=mask_.crop((min_y,min_x,max_y+1,max_x+1))
    #     img_=img_.crop((min_y,min_x,max_y+1,max_x+1))
    #     SAVEimg(img_,f"{filename_}.jpg",'rm')
    #     SAVEmask(mask_,f"{filename_}.png",'rm')
    #     count_+=1
    #     print(f'{dirname_}-rm-({count_}/{len_})',filename_)


    len_=len(os.listdir(composite_dir))
    count_=0
    listdir_=sorted([filename_ for filename_ in os.listdir(composite_dir) if '(' not in filename_])
    for filename_ in listdir_:
        temp_=filename_[:filename_.rfind('.')]
        if os.path.exists(f"{BASNetDS_dir_path}/im_aug_cm/{temp_}.jpg") \
          and os.path.exists(f"{BASNetDS_dir_path}/gt_aug_cm/{temp_}.png") :
            count_+=1
            # print(f'{dirname_}-cm-({count_}/{len_})',temp_)
            continue
        try:
          img_=Image.open(f"{composite_dir}/{filename_}")
        except:
          continue
        filename_=temp_
        filename_m=filename_
        filename_m=filename_m[:filename_m.rfind('_')]
        try:
          mask_=Image.open(f"{mask_dir}/{filename_m}.png")
        except:
          print(f"{dirname_}- {filename_m}")
          continue
        array_=np.array(mask_)
        min_x,min_y,max_x,max_y=10000,10000,0,0
        for i in range(array_.shape[0]):
            for j in range(array_.shape[1]):
                try:
                    if array_[i][j][0] != 0:
                        min_x,min_y,max_x,max_y=min(min_x,i),min(min_y,j),max(max_x,i),max(max_y,j)
                except:
                    if array_[i][j] != 0:
                        min_x,min_y,max_x,max_y=min(min_x,i),min(min_y,j),max(max_x,i),max(max_y,j)

        w_=(max_y-min_y+1)//10
        h_=(max_x-min_x+1)//10
        min_y-=random.randint(0,w_) # 左
        min_y=0 if min_y<0 else min_y
        min_x-=random.randint(0,h_) # 上
        min_x=0 if min_x<0 else min_x
        max_y+=random.randint(0,w_) # 右
        max_y=array_.shape[1] if max_y>array_.shape[1] else max_y
        max_x+=random.randint(0,h_) # 下
        max_x=array_.shape[0] if max_x>array_.shape[0] else max_x

        mask_=mask_.crop((min_y,min_x,max_y+1,max_x+1))
        img_=img_.crop((min_y,min_x,max_y+1,max_x+1))
        SAVEmask(mask_,f"{filename_}.png",'cm')
        SAVEimg(img_,f"{filename_}.jpg",'cm')
        count_+=1
        print(f'{dirname_}-cm-({count_}/{len_})',filename_)
    return len_



from IPython.display import clear_output

flag_=True
while flag_:
    total_=0
    total_+=run(dirlist_Hday2night)
    # clear_output()
    # print(f"total: {total_}")
    total_+=run(dirlist_HFlickr)
    # clear_output()
    # print(f"total: {total_}")
    total_+=run(dirlist_HCOCO)
    # clear_output()
    # print(f"total: {total_}")
    total_+=run(dirlist_HAdobe5k)
    # clear_output()
    clear_output()
    print(f"total: {total_}")
    flag_=False
    print("--Done--")
```
