```
import os
import re
from IPython.display import clear_output
import time
import random
from PIL import Image

BASNetDS_dir_path="/content/drive/Shareddrives/幻影捕手/AI班團專_幻影捕手/資料集/BASNetDS/im_aug_cm"

i=os.listdir(f'{BASNetDS_dir_path}_small')
len_=15000-i
dirlist_BASNetDS=os.listdir(BASNetDS_dir_path)
random.shuffle(dirlist_BASNetDS)
dirlist_BASNetDS=dirlist_BASNetDS[:len_]
i=i-1
for filename_ in dirlist_BASNetDS:
    filepath=f"{BASNetDS_dir_path}/{filename_}"
    img_=Image.open(filepath)
    new_filepath=f"{BASNetDS_dir_path}_small/{filename_}"
    img_.save(new_filepath)
    filename_=filename_.replace('jpg','png')
    filepath=f"{BASNetDS_dir_path.replace('im_aug_cm','gt_aug_cm')}/{filename_}"
    img_=Image.open(filepath)
    new_filepath=f"{BASNetDS_dir_path}_small/{filename_.replace('jpg','png')}"
    img_.save(new_filepath)
    i+=1
    print(f"{i}/{15000}")
print('Done')
```
